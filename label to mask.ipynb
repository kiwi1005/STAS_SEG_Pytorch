### 7.取出mask 並重新命名 
將程式碼內file_source 及 file_destination 替換成自己的路徑

```python
import os 
import shutil
```

```python
file_source = 'C:/Users/sheep/Desktop/TEST/label/' #所有資料夾的位置
file_destination = './data/masks/' #目的地位置 
a=0
for folder in os.listdir(file_source):
    a = folder[:-5]
    for file in os.listdir(os.path.join(file_source,folder)):
        if file == "label.png":
            print(file)
            s = "%08d" %int(a)
            file_oldname = os.path.join(file_source,folder,file)
            file_newname_newfile = os.path.join(file_destination,str(s)+'_mask.png')
            print(file_newname_newfile)
            shutil.copy(file_oldname, file_newname_newfile)
            a=int(a)
            a+=1
```
